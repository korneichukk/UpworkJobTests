```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://trustanalytica.com/ca/best-spa")
```

```python
items = driver.find_elements(By.CLASS_NAME, "sr-item")
len(items)
```

```python
items_data = {}

for item in items:
    item_details = item.find_element(By.CLASS_NAME, "sr-item-details")

    item_link_element = item_details.find_element(By.TAG_NAME, "a")
    item_link = item_link_element.get_attribute("href")
    item_link_text = item_link_element.text

    item_rating_element = item_details.find_element(
        By.CLASS_NAME, "sr-item-details-info-rating"
    )

    item_rating = item_rating_element.find_element(
        By.TAG_NAME, "span"
    ).text
    item_riviews = item_rating_element.find_element(
        By.TAG_NAME, "a"
    ).text

    description = item_details.find_element(By.TAG_NAME, "p").text

    contact_items = item.find_elements(
        By.CLASS_NAME, "sr-contact-item"
    )

    items_data[item_link] = {
        "link": item_link,
        "link_text": item_link_text,
        "rating": item_rating,
        "riviews": item_riviews,
        "description": description,
        "contract_items": [
            contact_item.text for contact_item in contact_items
        ],
    }
```

```python
import json
with open("items_data_analytics.json", "w") as f:
    json.dump(items_data, f)
```
